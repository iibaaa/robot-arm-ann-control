```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load file
relu_data = np.load("results/results_relu_model_relu.npy", allow_pickle=True).item()
tanh_data = np.load("results/results_tanh_model_tanh.npy", allow_pickle=True).item()
sigmoid_data = np.load("results/results_sigmoid_model_sigmoid.npy", allow_pickle=True).item()
```

```python
# Print Accuracies
relu_accuracy = relu_data["accuracy"]
tanh_accuracy = tanh_data["accuracy"]
sigmoid_accuracy = sigmoid_data["accuracy"]

# Plot All Accuracy in once
fig = plt.figure()
ax = fig.add_subplot(111)
ax.title.set_text('Accuracy Comparison for Activation Function')
ax.plot(relu_accuracy, label="relu")
ax.plot(tanh_accuracy, label="tanh")
ax.plot(sigmoid_accuracy, label="sigmoid")

ax.set_xlabel("Epoch")
ax.set_ylabel("Accuracy")
ax.legend()
ax.grid()
plt.show()
# Save Figure
fig.savefig("images/Accuracy.png")
```

```python
# Train Loses
relu_train_loss = relu_data["train_losses"]
tanh_train_loss = tanh_data["train_losses"]
sigmoid_train_loss = sigmoid_data["train_losses"]

# Plot all figures once
fig = plt.figure()
ax = fig.add_subplot(111)
ax.title.set_text('Train Loss Comparison for Activation Function')
ax.plot(relu_train_loss, label="relu")
ax.plot(tanh_train_loss, label="tanh")
ax.plot(sigmoid_train_loss, label="sigmoid")
ax.set_xlabel("Epoch")
ax.set_ylabel("Train Loss")
ax.legend()
ax.grid()
plt.show()
# Save Figure
fig.savefig("images/TrainLoss.png")

```

```python
# Test Loses
relu_test_loss = relu_data["test_losses"]
tanh_test_loss = tanh_data["test_losses"]
sigmoid_test_loss = sigmoid_data["test_losses"]

# Plot all figures once
fig = plt.figure()
ax = fig.add_subplot(111)
ax.title.set_text('Test Loss Comparison for Activation Function')
ax.plot(relu_test_loss, label="relu")
ax.plot(tanh_test_loss, label="tanh")
ax.plot(sigmoid_test_loss, label="sigmoid")
ax.set_xlabel("Epoch")
ax.set_ylabel("Test Loss")
ax.legend()
ax.grid()
plt.show()
# Save Figure
fig.savefig("images/TestLoss.png")
```

```python
# Print Tanh statistics
def display_activation_function(data, activation_function):
    fig = plt.figure()
    # Create 2 sub plot fi
    ax = fig.add_subplot(211)
    ax2 = fig.add_subplot(212)
    ax.title.set_text(f"{activation_function} Activation Function")
    
    ax.plot(data["train_losses"], label="train_losses")
    ax.plot(data["test_losses"], label="test_losses")
    ax.set_xlabel("Epoch")
    ax.grid()
    ax.legend()
    
    ax2.plot(data["accuracy"], label="accuracy")
    ax2.set_xlabel("Epoch")
    ax2.grid()
    ax2.legend()
    
    plt.show()
    
    fig.savefig(f"images/{activation_function}_activation_function.png")
    

display_activation_function(relu_data, "ReLu")
display_activation_function(tanh_data, "Tanh")
display_activation_function(sigmoid_data, "Sigmoid")
```

```python
# Diplay Comparison mean errors
relu_test_data = np.load("results/test_results_model_relu.npy", allow_pickle=True).item() 
tanh_test_data = np.load("results/test_results_model_tanh.npy", allow_pickle=True).item()
sigmoid_test_data = np.load("results/test_results_model_sigmoid.npy", allow_pickle=True).item()

fig = plt.figure()
ax = fig.add_subplot(111)

relu_avrg_error = np.mean(relu_test_data["error"], axis=1) * 1000
tanh_avrg_error = np.mean(tanh_test_data["error"], axis=1) * 1000
sigmoid_avrg_error = np.mean(sigmoid_test_data["error"], axis=1) * 1000

ax.title.set_text("Average Target Error Comparison for Activation Function")
ax.plot(relu_avrg_error, label="relu")
ax.plot(tanh_avrg_error, label="tanh")
ax.plot(sigmoid_avrg_error, label="sigmoid")

ax.set_xlabel("Test Number")
ax.set_ylabel("Average Target Error (mm)")
ax.legend(loc="upper left")
ax.grid()
plt.show()

fig.savefig("images/AverageTargetError.png")
```

```python
# Error Function
sigmoid_mae = np.load("results/results_sigmoid_model_sigmoid_mae.npy", allow_pickle=True).item()
sigmoid_mse = np.load("results/results_sigmoid_model_sigmoid_MSELoss.npy", allow_pickle=True).item()

fig = plt.figure()
ax = fig.add_subplot(211)

ax.plot(sigmoid_mae["train_losses"], label="train_losses")
ax.plot(sigmoid_mae["test_losses"], label="test_losses")
ax.set_xlabel("Epoch")
ax.set_ylabel("Loss")
ax.legend()
ax.grid()
ax.title.set_text("Sigmoid Activation Function with MAE Loss")

ax2 = fig.add_subplot(212)
ax2.plot(sigmoid_mae["accuracy"], label="accuracy")
ax2.set_xlabel("Epoch")
ax2.set_ylabel("Accuracy")
ax2.legend()
ax2.grid()

plt.show()

fig.savefig("images/SigmoidActivationFunctionWithMAELoss.png")

fig = plt.figure()
ax = fig.add_subplot(211)

ax.plot(sigmoid_mse["train_losses"], label="train_losses")
ax.plot(sigmoid_mse["test_losses"], label="test_losses")
ax.set_xlabel("Epoch")
ax.set_ylabel("Loss")
ax.legend()
ax.grid()
ax.title.set_text("Sigmoid Activation Function with MSE Loss")

ax2 = fig.add_subplot(212)
ax2.plot(sigmoid_mse["accuracy"], label="accuracy")
ax2.set_xlabel("Epoch")
ax2.set_ylabel("Accuracy")
ax2.legend()
ax2.grid()

plt.show()

fig.savefig("images/SigmoidActivationFunctionWithMSELoss.png")


# Print Accuracies
# Plot All Accuracy in once
fig = plt.figure()
ax = fig.add_subplot(111)
ax.title.set_text('Accuracy Comparison for Loss Function')
ax.plot(sigmoid_mae["accuracy"], label="mae")
ax.plot(sigmoid_mse["accuracy"], label="mse")

ax.set_xlabel("Epoch")
ax.set_ylabel("Accuracy")
ax.legend()
ax.grid()
plt.show()
# Save Figure
fig.savefig("images/AccuracyComparisonLossFunction.png")
```
